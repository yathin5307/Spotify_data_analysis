```python
df_tracks['duration']=df_tracks['duration_ms'].apply(lambda x:round(x/1000))
df_tracks.drop('duration_ms',inplace=True,axis=1)
```

```python
corr_df=df_tracks.drop(['key','mode'],axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df,annot=True,fmt=".1g",vmin=-1,vmax=1,center=0,cmap="inferno",linewidth=1,linecolor="Black")
heatmap.set_title("Correlation")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
sample_df=df_tracks.sample(int(0.04*len(df_tracks)))
print(len(sample_df))
```

Correlation between Loudness and Energy

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y="loudness",x="energy",color="c").set(title="Loudness vs Energy Correlation")
```

```python
sample_df.head(10)
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y="danceability",x="energy",color="b").set(title="Danceability vs Energy Correlation")
```

```python
genre=df_tracks.genre
plot1=sns.displot(genre,discrete=True,aspect=2,height=7,kind="hist",kde=True).set(title="No.of songs per Genre")
plot1.set_xticklabels(rotation=90)
```

```python
total_duration=df_tracks.duration
fig_dms=(18,7)
fig,ax=plt.subplots(figsize=fig_dms)
fig=sns.barplot(x=genre,y=total_duration,ax=ax,errwidth=False).set(title="Genre vs Duration")
plt.xticks(rotation=90)
```

```python
sns.set_style(style="darkgrid")
fig_dms=(17,8)
fig,ax=plt.subplots(figsize=fig_dms)
figure=sns.lineplot(x=genre,y=total_duration,ax=ax).set(title="Genre vs Duration")
plt.xticks(rotation=90)
```

```python

fig, ax = plt.subplots(figsize=(7, 10))
sns.color_palette("rocket",as_cmap=True)
sns.barplot(y="genre",x="danceability",ax=ax,data=df_tracks)

plt.xlabel("Danceability of the Genre",fontsize=22)
plt.ylabel("Genre",fontsize=22)
```
